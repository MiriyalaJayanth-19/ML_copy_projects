```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_data=sklearn.datasets.fetch_california_housing()
print(house_data)
```

```python
house_data_frame=pd.DataFrame(house_data.data, columns=house_data.feature_names)
```

```python
house_data_frame.head()
```

```python
house_data_frame['price']=house_data.target
house_data_frame.head()
```

```python
house_data_frame.describe()
```

```python
house_data_frame.shape
```

```python
house_data_frame.isnull().sum()
```

```python
correlation=house_data_frame.corr()
```

```python
print(correlation)
```

```python
plt.figure(figsize=(10,10))
sb.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Reds')
```

```python
x=house_data_frame.drop(['price'],axis=1)
y=house_data_frame['price']
print(x)
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,xtrain.shape,xtest.shape)
```

```python
model=XGBRegressor()
```

```python
model.fit(xtrain,ytrain)
```

```python
trainpred=model.predict(xtrain)
print(trainpred)
```

```python
print(metrics.r2_score(ytrain,trainpred))
```

```python
print(metrics.mean_absolute_error(ytrain,trainpred))
```

```python
testpred=model.predict(xtest)
```

```python
print(metrics.mean_absolute_error(ytest,testpred))
```

```python
print(metrics.r2_score(ytest,testpred))
```

```python
plt.scatter(ytrain,trainpred)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.show()
```
